fibG(n) = k fib(n) // h fib(n-1)

bisogna calcolare rG = fibG(n)//fibG(n-1)

quindi calcolare rG

```python
def fibogen(k : int , h : int, numfib : int):
    for (i) in range(1,(numfib + 1)):
        if (i == 1):
            print(f"k = {k}")
        elif (i == 2):
            print(f"h = {h}")
        else:
            tempnum = h
            h += k
            k = tempnum
            print(f"{h} dato da k {k} e h {h - k} ")
        
    print(f"Questo è il numero della sezione aurea {h/k}")
    print(f"Questo è rG  il quale si avvicina a 1 {(h/k)/((k)/(h-k))} dato da fibGen(n-1){(h/k)} e fibGen(n-2){((k)/(h-k))}")
            


k = int(input("Inserisci il valore k della tua sequenza di fibonacci"))
h = int(input("Inserisci il valore h della tua sequenza di fibonacci"))
numfib = int(input("Inserisci la lunghezza della sequenza di fibonacci "))

fibogen(k, h, numfib)
```

trovare il valore minimo di f(x) = (x-1)^2

[-3,3]

Come fare calcolare la derivata quindi f'(x) = 2(x-1) f'(x) = 0 quindi x = 1

Farla visualizzare in mathplotlib

Trovare il minimo e massimo tra [-3,3]





```python
import matplotlib.pyplot as math
import numpy as np

min = int(input("Inserisci il punto minimo della funzione"))
max = int(input("Inserisci il punto massimo della funzione"))

x = np.linspace(min, max, 100)
y = (x - 1)**2
y_derivata = 2 * (x - 1) 

math.plot(x, y, label='f(x) = (x-1)^2')
math.plot(x, y_derivata, label="f'(x) = 2(x-1)")
math.grid()

listaintersezioni = list() 
minimo = True
tempY = 0
for i in range(min, max + 1):
    y = (i - 1)**2
    y_derivata = 2 * (i - 1) 
    if y_derivata == y and minimo == True:
        listaintersezioni.append([i, y])
        math.plot(i, y, 'go', label="Valore minimo")
        minimo = False
    elif y_derivata == y and minimo == False:
        listaintersezioni.append([i, y])
        math.plot(i, y, 'b^', label = "Valore massimo intersezione derivata")
    elif y > tempY:
            massimo = [y, i]
            tempY = y
            
math.plot(massimo[1], massimo[0], 'rs', label="Valore massimo")
math.legend()        
listaintersezioni.append(massimo)  

print(listaintersezioni)
```
