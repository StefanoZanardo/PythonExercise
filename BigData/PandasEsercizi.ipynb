```python
import pandas as pd 
import pyspark as ps

df = pd.read_csv("ratings.csv" )
df.head(20)

```

Dobbiamo capire quante valutazioni ci sono nel dataset?

```python
numratings = df['rating'].count()

print(f"Il numero totale di valutazioni è: {numratings}")
```

Dobbiamo capire quanti film ci sono nel dataset

```python
nummovie = df['movieId'].nunique()
print(f"Il numero totale di libri è: {nummovie}")
```

Dobbiamo capire per ogni libro quante valutazioni ha ogni film

```python

df.groupby('movieId')['rating'].count().head(10)
```

Quali sono i 10 libri più valutati

```python
df.groupby('movieId')['rating'].count().sort_values(ascending=False).head(10)
```

Qual è la valutazione media per ogni libro

```python
df.groupby('movieId')['rating'].mean().head(10)
```

Quali sono i 10 libri con la valutazione più alta

```python
df.groupby('movieId')['rating'].mean().sort_values(ascending=False).head(10)
```

Chi sono i recensori più critici

```python
df.groupby('userId')['rating'].mean().sort_values(ascending=True).head(10).reset_index()
```
