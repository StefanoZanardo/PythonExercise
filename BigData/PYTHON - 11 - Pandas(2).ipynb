### ‚ùì Domanda 3
**Quali clienti hanno speso pi√π di 1000‚Ç¨ nel trimestre Marzo-Maggio?**

üëá Scrivi il codice qui sotto


````python
totali = df[(df['Data'].dt.month >= 3) & (df['Data'].dt.month <= 5)].groupby('Cliente').apply(lambda x: ((x['Prezzo Unitario'] * x['Quantit√†']) * (1 - x['Sconto'])).sum())
totali[totali> 1000]
````

<details>
<summary>üí° <strong>Mostra soluzione</strong></summary>

```python
df[(df['Data'].dt.month >= 3) & (df['Data'].dt.month <= 5)]\
  .groupby('Cliente')\
  .apply(lambda x: ((x['Prezzo Unitario'] * x['Quantit√†']) * (1 - x['Sconto'])).sum())\
  .query('> 1000')
```
