## üßº 2. Pulizia e trasformazione

- Sostituisci i valori mancanti in `Voto` con la media
- Mappa `Partecipazione`: Bassa=1, Media=2, Alta=3
- Crea una nuova colonna `Esito` (Passato se Voto ‚â• 18)


```python
#Sostituisci i valori mancanti in Voto con la media

#La colonna √® df['Voto'] e il metodo √® .fillna() che cerca i valori NaN e li sostituisce con df['Voto'].mean()
#ovvero la media della colonna Voto
df['Voto'] = df['Voto'].fillna(df['Voto'].mean())
df.head()
```

```python
#Mappa Partecipazione: Bassa=1, Media=2, Alta=3
df['Partecipazione'] = df['Partecipazione'].map({'Bassa': 1, 'Media': 2, 'Alta': 3})
df.head()
```

```python
#Un altro modo di scrivere le funzioni
lambda x: 'Passato' if x >= 18 else 'Non Passato'

def ciccio(x):
    if x >= 18:
        return "Passato"
    else:
        return "Non Passato"
```

```python
#Crea una nuova colonna Esito (Passato se Voto ‚â• 18)
df['Esito'] = df['Voto'].apply(lambda x: 'Passato' if x >= 18 else 'Non Passato')
df.head()
```

```python
#L'importanza dell'omogeinizzazione
df.describe(include='all')
```

## üìä 3. Analisi e aggregazioni

- Media voti per Classe e Materia
- Assenze medie per livello di partecipazione
- Studenti con pi√π di 3 assenze e voto < 22


```python
# Media voti per Classe e Materia
print(df.groupby(['Classe', 'Materia'])['Voto'].mean())
```

```python
#Assenze medie per livello di partecipazione
print(df.groupby('Partecipazione')['Assenze'].mean())
```

```python
#Studenti con pi√π di 3 assenze e voto < 22
df[(df['Assenze'] > 3) & (df['Voto'] < 22)]
```

## üîΩ 4. Ordinamento e selezione

- Ordina gli studenti per voto decrescente
- Mostra i 3 con pi√π ore di studio


```python
#Ordina gli studenti per voto decrescente
df.sort_values(by='Voto', ascending=False).head()
```

```python
#Mostra i 3 con pi√π ore di studio
df.sort_values(by='Ore Studio', ascending=False).head(3)
```

## ‚öôÔ∏è 5. Operazioni avanzate

- Crea colonna `Efficienza` = Voto / Ore Studio
- Crea una tabella pivot con:
    - indice: Materia
    - colonne: Classe
    - valori: media Voto


```python
#Crea colonna Efficienza = Voto / Ore Studio
df['Efficienza'] = df['Voto'] / df['Ore Studio']
df.head()
```

```python
#Crea tabella pivot
pivot = df.pivot_table(values='Voto', index='Materia', columns='Classe', aggfunc='mean')
pivot
```
